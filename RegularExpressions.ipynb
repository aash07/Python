```python
#Simple example on RegEx
# Display the lines which are containing 'how'

import re

fhand = open('myfile.txt')
for line in fhand:
    line = line.rstrip()
    if re.search('how', line):
        print(line)

fhand.close()
```

```python
#Display only those lines which are starting with 'how'
import re
hand = open('myfile.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^how', line):
        print(line)
  
hand.close()
```

```python
# Write a program to extract lines starting with the word 'From' (or 'from') 
# and ending with 'edu'
# Input file is: mbox-short.txt

import re
fhand = open('mbox-short.txt')
count=0
for line in fhand:
    count+=1
    line = line.rstrip()
    pattern = '^[Ff]rom.*edu$'
    if re.search(pattern, line):
        print("Line number:", count, "Line:", line)

fhand.close()
```

```python
# Extract all the lines ending with any digit

import re
fhand = open('mbox-short.txt')

for line in fhand:
    line = line.rstrip()
    pattern = '[0-9]$'
    if re.search(pattern, line):
        print(line)

fhand.close()
```

```python
# line must be started with anything other than
# lower-case alphabets and digits.

import re
fhand = open('mbox-short.txt')

for line in fhand:
    line = line.rstrip()
    pattern = '^[^a-z0-9]'
    if re.search(pattern, line):
        print(line)

fhand.close()
```

```python
#line should start with capital letters, followed by zero or more characters, 
# but must end with any digit

import re
fhand = open('mbox-short.txt')

for line in fhand:
    line = line.strip()
    pattern = '^[A-Z].*[0-9]$'
    if re.search(pattern, line):
        print(line)

fhand.close()
```

```python
# Demo of findall() method
# Extract email ID from a given string

import re
s = 'A message from csev@umich.edu to cwen@iupui.edu about meeting @2PM'

lst = re.findall('\S+@\S+', s)
print(lst)

#Extract only the domain names

for email in lst:
    x= email.find('@')
    print(email[x+1:])
```

```python
# Get a list of all the email IDs from a file, by removing extra leading and 
# trailing characters like > ; etc. And also to remove non-email patterns like
# xyz@localhost etc.

import re
fhand = open('mbox-short.txt')
ls=[]
for line in fhand:
    line = line.strip()
    x = re.findall('[a-zA-Z0-9]\S+@\S+[a-zA-Z][.][a-zA-Z]+', line)
    if len(x) > 0:
        ls.extend(x)
        
print(ls)


```

```python
# Extract only the revision numbers from the hyperlinks given in the file

import re
hand = open('mbox-short.txt')
ls=[]
for line in hand:
    line = line.strip()
    x = re.findall('^Details:.*rev=([0-9]+)', line)
    if len(x)>0:
        ls.extend(x)

print(ls)
```

```python
import re
hand = open('mbox-short.txt')
ls1=[]
ls2=[]
for line in hand:
    line = line.strip()
    mails_from_pattern = '^From .* ([0-9][0-9]):'
    mails_received_pattern = '^[a-zA-Z]{3}, .* ([0-9][0-9]):'
    x=re.findall(mails_from_pattern, line)
    if len(x)>0:
        ls1.extend(x)
    
    y= re.findall(mails_received_pattern,line)
    if len(y)>0:
        ls2.extend(y)

ls3=ls1+ls2
print(ls3)

freq=[]
for i in ls3:
    freq.append((i,ls3.count(i)))

hist=list(set(freq))
hist.sort(key=lambda x:(x[1],x[0]),reverse=True)
print(hist)
```
